```python
from reportlab.lib.pagesizes import letter
from reportlab.lib.pagesizes import landscape
from reportlab.pdfgen import canvas
import fitz  # Importar PyMuPDF

# Tamao de una tarjeta de crdito en puntos (1 pulgada = 72 puntos)
width, height = 85.60 * 72, 53.98 * 72

# Crear un archivo PDF
c = canvas.Canvas("tarjeta_de_credito.pdf", pagesize=(width, height))

# Dibuja un rectngulo del tamao de la tarjeta de crdito
c.rect(0, 0, width, height)


data = { 
    "CustomerID": [101],
    "Name": ["John Doe"],
    "Email": ["john@example.com"],
    "Phone Number": ["123-456-7890"],
    "Mailing Address": ["123 Main St"],
    "City": ["City A"],
    "Province": ["Province X"],
    "Country": ["Country 1"]
}


# Definir una lista de elementos de texto con sus propiedades
text_elements = [
    {"text": f"Customer ID: {data['CustomerID']}", "size": 72, "x": width * 0.1, "y": height * 0.40},
    {"text": f"Name: {data['Name']}", "size": 72, "x": width * 0.1, "y": height * 0.42},
    {"text": f"Email: {data['Email']}", "size": 72, "x": width * 0.1, "y": height * 0.44},
    {"text": f"Phone Number: {data['Phone Number']}", "size": 72, "x": width * 0.1, "y": height * 0.46},
    {"text": f"Mailing Address: {data['Mailing Address']}", "size": 72, "x": width * 0.1, "y": height * 0.48},
    {"text": f"City: {data['City']}", "size": 72, "x": width * 0.1, "y": height * 0.50},
    {"text": f"Province: {data['Province']}", "size": 72, "x": width * 0.1, "y": height * 0.52},
    {"text": f"Country: {data['Country']}", "size": 72, "x": width * 0.1, "y": height * 0.54},
]

# Agregar los elementos de texto al PDF usando un bucle
for element in text_elements:
    c.setFont("Helvetica", element["size"])
    c.drawString(element["x"], element["y"], element["text"])

# Guardar el archivo PDF
c.save()
```

```python
# Importar las bibliotecas necesarias
import pandas as pd
import qrcode
import os
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from reportlab.lib.utils import ImageReader
import matplotlib.pyplot as plt

# DataFrame con informacin de clientes
data = {
    "CustomerID": [101, 102, 103, 104],
    "Name": ["John Doe", "Jane Smith", "Michael Johnson", "Emily Brown"],
    "Email": ["john@example.com", "jane@example.com", "michael@example.com", "emily@example.com"],
    "Phone Number": ["123-456-7890", "987-654-3210", "555-123-4567", "888-999-0000"],
    "Mailing Address": ["123 Main St", "456 Elm St", "789 Oak Ave", "555 Pine Rd"],
    "City": ["City A", "City B", "City C", "City D"],
    "Province": ["Province X", "Province Y", "Province Z", "Province W"],
    "Country": ["Country 1", "Country 2", "Country 3", "Country 4"]
}

df_customers = pd.DataFrame(data)

# Ruta donde se guardarn los archivos PDF
folder_path = "C:\\Users\\Admin\\Desktop\\Portfolio\\PDF_Generation\\Output"

# Comprueba si la carpeta existe y crala si no
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Ruta de la imagen a superponer en la parte superior derecha
image_path = "C:\\Users\\Admin\\Desktop\\Portfolio\\PDF_Generation\\Input\\Chickadee.png"

# Tamao de la imagen a superponer en puntos
image_width, image_height = 100, 100

# Genera los cdigos QR y archivos PDF para cada cliente
for index, customer in df_customers.iterrows():
    customer_id = customer["CustomerID"]
    customer_name = customer["Name"]
    customer_email = customer["Email"]
    customer_city = customer["City"]
    customer_province = customer["Province"]

    full_qr_filename = os.path.join(folder_path, f"QR_Customer_{customer_id}.png")

    # Genera el cdigo QR y gurdalo en un archivo
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    
    # Combina la informacin en un solo texto
    customer_info = f"CustomerID: {customer_id}\nName: {customer_name}\nEmail: {customer_email}\nCity: {customer_city}\nProvince: {customer_province}"
    qr.add_data(customer_info)
    qr.make(fit=True)
    img = qr.make_image(fill_color="black", back_color="white")
    img.save(full_qr_filename)

    # Tamao de una tarjeta de crdito en puntos (1 pulgada = 72 puntos)
    width, height = 85.60 * 72, 53.98 * 72

    # Crea un archivo PDF con el cdigo QR del cliente
    pdf_file = os.path.join(folder_path, f"QR_Customer_{customer_id}.pdf")
    c = canvas.Canvas(pdf_file, pagesize=(width, height))

    # Dibuja un rectngulo del tamao de la tarjeta de crdito
    c.rect(0, 0, width, height)

    # Definir una lista de elementos de texto con sus propiedades
    text_elements = [
        {"text": f"Customer ID: {customer_id}", "size": 72, "x": width * 0.1, "y": height * 0.40},
        {"text": f"Name: {customer_name}", "size": 72, "x": width * 0.1, "y": height * 0.42},
        {"text": f"Email: {customer_email}", "size": 72, "x": width * 0.1, "y": height * 0.44},
        {"text": f"City: {customer_city}", "size": 72, "x": width * 0.1, "y": height * 0.46},
        {"text": f"Province: {customer_province}", "size": 72, "x": width * 0.1, "y": height * 0.48},
    ]

    # Agregar los elementos de texto al PDF usando un bucle
    for element in text_elements:
        c.setFont("Helvetica", element["size"])
        c.drawString(element["x"], element["y"], element["text"])

    # Cargar la imagen para superponer
    qr_image = ImageReader(full_qr_filename)
    c.drawImage(qr_image, 80, 80, width * 0.3, height * 0.3)

    # Cargar la imagen para superponer en la parte superior derecha
    c.drawImage(image_path, width - image_width,
                height - image_height,
                width=image_width,
                height=image_height
                )

    # Guardar el archivo PDF
    c.save()

    # Mostrar los primeros 3 cdigos QR
    if index < 3:
        plt.imshow(img)
        plt.title(f"QR code for Customer {customer_id}")
        plt.show()

print("Proceso completado.")
```
